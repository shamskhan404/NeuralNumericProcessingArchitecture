```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
def generate_data(n=10000):
    x1 = np.random.randint(1, 20, n)
    x2 = np.random.randint(1, 20, n)
    
    y = x1 + x2
    
    X = np.stack([x1, x2], axis=1)
    y = y.reshape(-1, 1)
    
    return torch.FloatTensor(X), torch.FloatTensor(y)
```

```python
# 2. Model Definition

class SimpleMLP(nn.Module):
    def __init__(self):
        super(SimpleMLP, self).__init__()
        
        self.model = nn.Sequential(
            nn.Linear(2, 16),   # Input → Dense
            nn.ReLU(),          # Activation
            nn.Linear(16, 1)    # Dense → Output
        )
    
    def forward(self, x):
        return self.model(x)
```

```python
model = SimpleMLP()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# Generate data
X, y = generate_data(10000)
```

```python
# 4. Training Loop
epochs = 100

for epoch in range(epochs):
    model.train()
    
    optimizer.zero_grad()
    
    outputs = model(X)
    loss = criterion(outputs, y)
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
# 5. Testing
model.eval()

test_data = torch.FloatTensor([[5, 7], [10, 3], [100, 200]])

predictions = model(test_data)

print("\nTest Predictions:")
for inp, pred in zip(test_data, predictions):
    print(f"{inp.tolist()} → Predicted: {pred.item():.2f}, Actual: {sum(inp.tolist())}")
```
